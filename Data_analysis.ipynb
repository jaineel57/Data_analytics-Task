**Question - Standardise the mail_sent_date column values to YYYY-MM-DD standard date format**

```python
df_email['mail_sent_date'] = pd.to_datetime(df_email['mail_sent_date'])
df_email.head()
```

```python
df_email['year'] = df_email['mail_sent_date'].dt.year
df_email.head()
```

**Question - In which Year maximum number of emails were sent to doctors?**

```python
result = df_email.groupby('year')['year'].count()
result
```

**In th year 2014 is the answer**

```python
df_fine[' online_fine '] = df_fine[' online_fine '].str.replace('-','0')
df_fine[' cash_fine '] = df_fine[' cash_fine '].str.replace('-','0')
df_fine.head()
```

```python
df_fine[' cash_fine '].isnull().values.any()
```

**Question - Write a query to get the highest total fine and lowest total fine paid per state**

```python
output2 = df_fine.groupby('state').sum()['Total_fine']
output2.sort_values(ascending=False)
```

```python
output2.sort_values()
```

**Question - Write a query to get the least 2 states that had to pay minimum 'online' fine**

```python
output3 = df_fine.groupby('state').sum()[' online_fine ']
output3.sort_values().tail(2)
```

**The above code output is the answer**

**Question- Get the top 5 average fine paid per state and category**

```python
output5 = df_fine.groupby('state').mean()['Total_fine']
output5.sort_values(ascending=False).head(5)
```

**The above output of the code is for per state**

```python
output5 = df_fine.groupby('category').mean()['Total_fine']
output5.sort_values(ascending=False).head(5)
```
